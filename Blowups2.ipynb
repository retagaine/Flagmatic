```python
from flagmatic.all import *
```

```python
def fn(r, n, subgraph_order):
    str_range = map(lambda l: str(l), range(1, n+1))
    edges = Combinations(str_range * 2, r).list()
    edges = map(lambda l: ("").join(l), edges)

    highest_densities = dict()
    count = 0

    for i in range(1, len(str_range)+1):
        picked_edges = Combinations(edges, i).list()
        for p in picked_edges:
            p = ("").join(p)
            s = "{}:{}".format(n, p)
            try:
                densities = FourGraphBlowupConstruction(s).subgraph_densities(subgraph_order)
                for d in densities:
                    if d[0].ne > 0:
                        if d[0] not in highest_densities:
                            highest_densities[d[0]] = (d[1], s)
                        else:
                            if highest_densities[d[0]][0] < d[1]:
                                highest_densities[d[0]] = (d[1], s)
                count += 1
            except:
                continue
            if count % 1000 == 0:
                print(highest_densities)
                with open("certs/4-graph_densities.txt", "a") as f:
                    f.write(str(highest_densities))
                    f.write('\n')
                    f.close()
        print(highest_densities)
        with open("certs/4-graph_densities.txt", "a") as f:
            f.write(str(highest_densities))
            f.write('\n')
            f.close()
    print(highest_densities)
    with open("certs/4-graph_densities.txt", "a") as f:
        f.write(str(highest_densities))
        f.write('\n')
        f.close()
```
